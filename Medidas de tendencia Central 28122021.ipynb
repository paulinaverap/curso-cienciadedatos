## Media (Promedio), Mediana y la Moda

```python
import pandas as pd
pd.options.display.max_columns = None
import matplotlib.pyplot as plt
plt.style.use("seaborn-dark")
```

```python
df = pd.read_csv('cars.csv') # Data de automoviles usados
```

```python
df.head()
```

```python
df['price_usd'].mean()
```

```python
# Nos interesa analizar los atipicos tambien es decir cual es el auto con mayor precio
# cual de todos es el barato
```

```python
df['price_usd'].median()
# con la mediana puedo indentificar que existen autos costosos a pesar de ser usados
# El promedio esta lejano a la media
```

```python
# Vamos a graficar la columna price_usd
df['price_usd'].plot.hist(bins=20, edgecolor="black") # 20 segmentaciones o intervalos de valores
plt.show()
```

```python
# la mayoria de los autos estan entre 0 a 10000, pero tenemos varios casos atipicos
```

```python
import seaborn as sns
```

```python
sns.displot(df, x='price_usd', hue ='manufacturer_name')
plt.show()
```

```python
# pasamos a una visualizacion mas sencilla
sns.displot(df, x='price_usd', hue='engine_type', multiple='stack')
plt.show()
```

```python
# Aun sigo sin poder los autos electricos, puedo intuir que hay pocos autos electricos
# conteo tambien es un estadistico descriptivo
# cuantos autos hay por cada tipo de maquina de acuerdo a su motor (engine_type)?
```

```python
# busquemos un auto que sea de marca audi y de la clase Q7
Q7_df = df[(df['manufacturer_name']=='Audi') & (df['model_name']=='Q7')]
```

```python
#Q7_df.iloc[:]
Q7_df
```

```python
# Modelo por año
sns.histplot(Q7_df, x='price_usd', hue='year_produced')
plt.show()
```

Observaciones
La mayoria de los autos Q7 que estan en este catalogo
tienden a tener precios bajos entre 10000 a 20000, los colores claros hacen referencia
a los modelos mas antiguos del 2006, 2007, 2008
el auto mas antiguo va depreciando su valor


```python
# busquemos un auto que sea de marca Volkswagen modelo Golf
golf_df = df[(df['manufacturer_name']=='Volkswagen') & (df['model_name']=='Golf')]
```

```python
golf_df
```

```python
# Modelo por año
sns.histplot(golf_df, x='price_usd', hue='year_produced')
plt.show()
```

Se observa que los modelos Golf producidos entre los años 1976 y 1990 son los más económicos llegando a costar hasta 2500 USD. A su vez se observa que los modelos más actuales se encuentran en un rango de precio que va de los 2500 a los 15000 USD. Como la muestra es de 707 autos, para chequear si efectivamente los autos producidos hasta el 1990 son los mas economicos, realizé un segundo gráfico sólo con aquellos autos producidos hasta 1990.

```python
old_golf_df = golf_df[(golf_df['year_produced']<=1990)]
```

```python
old_golf_df
```

```python
sns.histplot(old_golf_df, x='price_usd', hue='year_produced')
plt.show()
```

Efectivamente los autos producidos hasta 1990 tienen un valor igual o inferior a 2500 USD.
