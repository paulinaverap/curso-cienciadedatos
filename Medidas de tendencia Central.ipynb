```python
import seaborn as sns
```

```python
# busquemos un auto que sea de marca audi y de la clase Q7
Q7_df = df[(df['manufacturer_name']=='Audi') & (df['model_name']=='Q7')]
```

```python
# busquemos un auto que sea de marca Volkswagen y de la clase Golf
golf_df = df[(df['manufacturer_name']=='Volkswagen') & (df['model_name']=='Golf')]
```

```python
golf_df
```

```python
sns.histplot(golf_df, x='price_usd', hue='year_produced', multiple='stack')
plt.show()
```

Existen 707 Volkswagen Golf en la muestra, de los cuales, cerca de 370 autos tienen un valor inferior o igual a 2500 USD. Como no se puede apreciar bien los colores del gráfico pareciera ser que la mayoría de esos autos fueron producidos entre los años 1976 y 1990, para comprobar esto último se seleccionaran sólo aquellos autos producidos antes del 1990. 

```python
old_golf_df = golf_df[(golf_df['year_produced']<=1990)]
```

```python
old_golf_df
```

```python
sns.histplot(old_golf_df, x='price_usd', hue='year_produced', multiple='stack')
plt.show()
```

Efectivamente los autos producidos en fecha anterior a 1990 tienen un valor menor o igual a 2500 USD, pero no son cerca de 370 como se visualizaban en el gráfico anterior, sino que solo son 130 autos correspondientes a la epoca, por lo que el primer gráfico tiende a error visual por la gran variedad de años de producción de los autos.
